<h2 style="color:#f7ce5b; font-family: Arial, sans-serif; font-size: 40px; margin-top: 50px; border-bottom: 2px solid #f7ce5b;">
Library
</h2>

```python
import pandas as pd 
import numpy as np 

from tensorflow import keras
from tensorflow.keras import layers, models

from sklearn.model_selection import train_test_split
```

<h2 style="color:#f7ce5b; font-family: Arial, sans-serif; font-size: 40px; margin-top: 50px; border-bottom: 2px solid #f7ce5b;">
Data
</h2>

```python
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
```

```python
display(train)
```

We create our X_train and Y_train. 

```python
Y_train = train['label']
X_train = train.iloc[:, 1:]
```

We make them scale by divide with 255. And then the second line of code reshapes the input data from a 2D array (where each row represents an image with 784 pixels) to a 4D tensor with shape.

```python
X_train = X_train / 255.0
X_train = np.array(X_train)
X_train = X_train.reshape(-1, 28, 28, 1)

X_test = test/255
X_test = np.array(X_test)
X_test = X_test.reshape(-1, 28, 28, 1)
```

We create a validation set for our model.

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, Y_train, test_size=0.2, random_state=13)
```

<h2 style="color:#f7ce5b; font-family: Arial, sans-serif; font-size: 40px; margin-top: 50px; border-bottom: 2px solid #f7ce5b;">
Model
</h2>

```python
model = models.Sequential(
    [
        layers.Conv2D(20, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(32, (2, 2), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(units=200, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(units=10, activation='sigmoid')
    ]
)
model.summary()
```

We compile and fit the model with our data.

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, validation_data=(X_val, y_val))
```

```python
model.save('./model/model.h5')
```

<h2 style="color:#f7ce5b; font-family: Arial, sans-serif; font-size: 40px; margin-top: 50px; border-bottom: 2px solid #f7ce5b;">
Prediction
</h2>

```python
y_pred = model.predict(X_test)
predict = np.argmax(y_pred, axis = -1)
```

<h2 style="color:#f7ce5b; font-family: Arial, sans-serif; font-size: 40px; margin-top: 50px; border-bottom: 2px solid #f7ce5b;">
Submission
</h2>

```python
submission = pd.read_csv('./data/sample_submission.csv', index_col=0)
```

```python
submission['Label'] = list(predict)
submission.to_csv('./results/result.csv')
```
